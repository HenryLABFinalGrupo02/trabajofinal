```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
business = pd.read_json("./data/business.json", lines=True)
```

```python
business.head()
```

```python
business = business.set_index('business_id')
```

## ATTRIBUTES

```python
attributes = pd.json_normalize(business['attributes'])
```

```python
attributes.set_index(business.index, inplace=True)
```

```python
attributes['business_id'] = attributes.index
```

```python
attributes.reset_index(drop=True, inplace=True)
```

```python
attributes
```

```python
attributes.shape
```

```python
attributes.to_csv('./data/attributes.csv', index=False)
```

## OPEN HOURS

```python
openhours = pd.json_normalize(business['hours'])
```

```python
openhours.set_index(business.index, inplace=True)
openhours['business_id'] = openhours.index
```

```python
openhours.to_csv('./data/openhours.csv', index=False)
```
