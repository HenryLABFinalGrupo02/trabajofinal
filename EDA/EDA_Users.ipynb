```python
def valores_unicos(data): 
  '''
  Esta función detecta los valores únicos en el dataset 

  :: param data: DataFrame de pandas 

  Return: 
  ---------
  Devuelve una lista con los valores único en cada una de las columnas del DatFrame
  '''
  dict_val_unicos = {}

  columnas = data.columns 
  for columna in columnas: 
    valores_unicos = data[columna].unique()
    dict_val_unicos[columna] = valores_unicos

  return dict_val_unicos
```

```python
def get_dtypes(dataset):
  '''
  This function gets the data types present on each column of a dataset.

  :: param dataset: Pandas or Koalas DataFrame

  Return:
  ---------
  Returns a dictionary that contains the data types for each column of a dataset.
  '''
  # npds = dataset.values
  cols = dataset.columns
  types_dict = {}
  for col in cols:
    types_dict[col] = set()
    for value in dataset[col]:
      types_dict[col].add(type(value))
  return types_dict
```
