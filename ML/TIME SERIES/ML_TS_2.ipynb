```python
import pandas as pd
import plotly.express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objects as go
import numpy as np
pd.options.plotting.backend = 'plotly'
```

```python
train_df = pd.read_csv('./data/train_ts.csv', index_col='month', parse_dates=True)
forecast_df = pd.read_csv('./data/forecast_ts.csv', index_col='month', parse_dates=True)
```

```python
for i in train_df.columns.tolist():
    fig1 = px.line(train_df[i])
    fig1.update_layout(title='Actual')
    fig1.update_traces(line_color='purple', name='Actual')

    fig2 = px.line(forecast_df[i])
    fig2.update_layout(title='Forecast')
    fig2.update_traces(line_color='seagreen', name='Forecast')

    fig = go.Figure(data = fig1.data + fig2.data)
    fig.update_layout(title=i)
    iplot(fig)
```
